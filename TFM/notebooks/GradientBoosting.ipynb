En este notebook a probar Gradient Boosting. Al igual que random forest es una combinación de otros modelos de predicción más débiles, casi siempre árboles de decisión. A la salida de cada árbolde decisión se le aplica un corrector. Este corrector se va ajustando por descenso de gradiente. 

Para empezar vamos a probar el algoritmo en bruto, para ver su rendimiento.


```python
trainx,trainy,testx,testy = PreparaDatosEntrenamientoTest(dataSet)
```

```python
clfGradientBoosting=GradientBoostingClassifier()
prediccion=Clasifica(clfGradientBoosting,trainx,trainy,testx)
CurvaRocAnalisis(testy,prediccion)
pd.crosstab(testy['default payment next month'],prediccion, rownames=['Default test'], colnames=['Default predicho'])
```

Vemos que el mejor resultado lo obtenemos con el número de hojas 14. Ahora vamos a tunear los parámetros utilizando un grid. Para ello vamos a optimizar los parámetros propios de los árboles de decisión, como max_depth, n_stimators, min_samples_leaf, como hicimos en random forest, pero también vamos a optimizar un parámetro propio de Grandient boosting. Esa es la tasa de aprendizaje, gradient boosting va configurando y añadiendo nuevos árboles secuencialmente, para corregir el resultado de los anteriores. A la corrección se le suele aplicar una tasa, la cual vamos a optimizar. Se llama shrinkage. Es usual utilizar factores bajos.

```python
X = dataSet.sample(5000).ix[:,:(len(dataSet.sample(5000).columns)-1)]
Y = dataSet.sample(5000).ix[:,(len(dataSet.sample(5000).columns)-1):]


learningRate=(0.0001,0.001,0.01,0.1,0.2)

param_grid = {'learning_rate':learningRate}
grid_search = GridSearchCV(GradientBoostingClassifier(max_depth=50,n_estimators=1000,min_samples_leaf=50,random_state=0), param_grid, cv=14, scoring='roc_auc')
grid_search.fit(X, Y['default payment next month'])
print(grid_search.best_params_)
```
